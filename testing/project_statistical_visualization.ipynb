```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('cleaned_grad_cafe_admissions.csv')

# top 10
freq = data['institution'].value_counts().nlargest(10)

plt.title('Top 10 most popular school')
plt.bar(freq.index, freq.values)
plt.xticks(rotation=90)
plt.xlabel('Institution')
plt.ylabel('Frequency')
plt.show()
```

```python
df = pd.read_csv('cleaned_grad_cafe_admissions.csv')


# Count the frequency of each institution in the 'institution' column
freq = df['institution'].value_counts()

# Get the top 10 most frequent institutions
top10 = freq.head(10)

# Create a dictionary to store the average GPA and GRE for each institution
data = {}

# Loop through the top 10 institutions and compute the average GRE and GPA for each one
for institution in top10.index:
    institution_df = df.loc[df['institution'] == institution]
    avg_gpa = institution_df['gpa'].mean()
    avg_gre = institution_df['gre'].mean()
    print(f"Institution: {institution}, Average GPA: {avg_gpa:.2f}, Average GRE: {avg_gre:.2f}")
    data[institution] = {'avg_gpa': avg_gpa, 'avg_gre': avg_gre}

# Convert the data dictionary to a DataFrame
plot_df = pd.DataFrame(data).T
```

```python
plot_df.plot(kind='line',y=['avg_gre'])
plt.xticks(range(len(top10.index)), top10.index, rotation=90)
plt.xlabel('Institution')
plt.ylabel('Average Value')
plt.title('Average GRE for Top 10 Institutions')
plt.show()
```
