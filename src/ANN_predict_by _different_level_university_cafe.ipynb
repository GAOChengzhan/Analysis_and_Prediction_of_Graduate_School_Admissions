```python
# Created by Weirong Dong
# Use GPA GRE GRE_W to predict adimission possibility for different level univ
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# read excel file
df = pd.read_excel("rank_processed.xlsx")

# group data based on institution
groups = df.groupby('institution')
#initialize
arrays = [np.array([]) for _ in range(10)]
labels = [np.array([]) for _ in range(10)]

# store the data
for group_name, group_data in groups:
    h_values = group_data['gpa'].values
    i_values = group_data['gre'].values
    j_values = group_data['gre_w'].values
    k_values = group_data['Result'].values
    arrays[group_name] = np.column_stack((h_values, i_values, j_values))
    labels[group_name] = k_values

# splite the data
test_size = 0.2  
random_state = 42  
train_arrays = []
train_labels = []
test_arrays = []
test_labels = []
for i in range(9):
    if arrays[i].size > 0:
        X_train, X_test, y_train, y_test = train_test_split(arrays[i], labels[i], test_size=test_size, random_state=random_state)
        train_arrays.append(X_train)
        train_labels.append(y_train)
        test_arrays.append(X_test)
        test_labels.append(y_test)

```

```python
import openpyxl
import torch
import numpy as np
from torch.utils.data import DataLoader, random_split
from sklearn.model_selection import train_test_split
import torch.nn as nn
import torch.optim as optim

# Define the neural network class
class NeuralNetwork(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc1 = nn.Linear(input_size, 50)
        self.fc2 = nn.Linear(50, 100)
        self.fc3 = nn.Linear(100, 100)
        self.fc4 = nn.Linear(100, 50)
        self.fc5 = nn.Linear(50, output_size)
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = torch.relu(self.fc3(x))
        x = self.dropout(x)
        x = torch.relu(self.fc4(x))
        x = self.dropout(x)
        x = torch.sigmoid(self.fc5(x))
        #x = torch.softmax(self.fc5(x), dim=1)
        return x
    
    def train(self, train_data, train_labels, epochs=1000, batch_size=8192, lr=0.05):
        loss_fn = torch.nn.CrossEntropyLoss()
        optimizer = optim.Adam(self.parameters(), lr=lr)
        #optimizer = torch.optim.SGD(self.parameters(), lr=lr)
        train_dataset = torch.utils.data.TensorDataset(train_data, train_labels)
        train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size)
        for epoch in range(epochs):
            for batch_x, batch_y in train_loader:
                optimizer.zero_grad()
                outputs = self.forward(batch_x)
                loss = loss_fn(outputs, batch_y)
                loss.backward()
                optimizer.step()
                if (epoch + 1) % print_every == 0:
                    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
    
    def predict(self, test_data):
        outputs = self.forward(test_data)
        _, y_pred = torch.max(outputs, dim=1)
        return y_pred.numpy()
    
```

```python
for i in range(9):
    train_arrays[i] = torch.tensor(train_arrays[i], dtype=torch.float32)
    train_labels[i] = torch.tensor(train_labels[i], dtype=torch.long)
    test_arrays[i] = torch.tensor(test_arrays[i], dtype=torch.float32)
    test_labels[i] = torch.tensor(test_labels[i], dtype=torch.long)
    print_every = 15
# Create and train the neural network
    #nn = NeuralNetwork(input_size=3, output_size=2)
    #nn.train(train_arrays[i], train_labels[i], epochs=15)
nn = NeuralNetwork(input_size=3, output_size=2)
nn.train(train_arrays[0], train_labels[0], epochs=15)
nn.train(train_arrays[1], train_labels[1], epochs=15)
nn.train(train_arrays[2], train_labels[2], epochs=15)
nn.train(train_arrays[3], train_labels[3], epochs=15)
nn.train(train_arrays[4], train_labels[4], epochs=15)
nn.train(train_arrays[5], train_labels[5], epochs=15)
nn.train(train_arrays[6], train_labels[6], epochs=15)
nn.train(train_arrays[7], train_labels[7], epochs=15)
nn.train(train_arrays[8], train_labels[8], epochs=15)
#Make predictions on the testing set and calculate accuracy
y_pred = nn.predict(test_arrays[0])
accuracy = np.mean(y_pred == test_labels[0].numpy())
print("Group 0 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[1])
accuracy = np.mean(y_pred == test_labels[1].numpy())
print("Group 1 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[2])
accuracy = np.mean(y_pred == test_labels[2].numpy())
print("Group 2 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[3])
accuracy = np.mean(y_pred == test_labels[3].numpy())
print("Group 3 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[4])
accuracy = np.mean(y_pred == test_labels[4].numpy())
print("Group 4 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[5])
accuracy = np.mean(y_pred == test_labels[5].numpy())
print("Group 5 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[6])
accuracy = np.mean(y_pred == test_labels[6].numpy())
print("Group 6 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[7])
accuracy = np.mean(y_pred == test_labels[7].numpy())
print("Group 7 " + f"Accuracy: {accuracy}")
y_pred = nn.predict(test_arrays[8])
accuracy = np.mean(y_pred == test_labels[8].numpy())
print("Group 8 " + f"Accuracy: {accuracy}")
```
